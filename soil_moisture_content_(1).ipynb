```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y2_train, y2_test = train_test_split(X, y2, test_size = 0.20, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
l2 = LinearRegression()
l2.fit(X_train,y2_train)
```

```python
accuracy2 = l2.score(X_test,y2_test)
print(accuracy2*100,'%')
```

```python
from sklearn.ensemble import RandomForestRegressor
accuracy_score = []
no_tree = [10,70,90,100,500]
max_depth_array = [2,3,4,5,6]
for i in range(len(no_tree)):
    for j in max_depth_array:
        regressor = RandomForestRegressor(n_estimators = no_tree[i], max_depth=j, random_state = 0)
        regressor.fit(X_train, y2_train)
        accuracy = regressor.score(X_test,y2_test)
        accuracy_score.append((accuracy*100,no_tree[i],j))
```

```python
max_value = 0
for i in accuracy_score:
    print(i)
    if(max_value<i[0]):
        max_value = i[0]
        max_ele = i
print("max accuracy obtain with parameters n_estimators as {0} and max_depth as {1} ".format(max_ele[1],max_ele[2]))
```

```python
plt.figure(figsize=(5,5))
for j in max_depth_array:
    r2RFR = []
    notree =[]
    for i in accuracy_score:
        if(i[2]==j):
            r2RFR.append(i[0])
            notree.append(i[1])
    plt.plot(notree,r2RFR)
    plt.xlabel("no of tree")
    plt.ylabel("R2_Score")
plt.legend(max_depth_array)
    
    
```

```python
plt.figure(figsize=(5,5))
for j in no_tree:
    r2RFR = []
    notree =[]
    for i in accuracy_score:
        if(i[1]==j):
            r2RFR.append(i[0])
            notree.append(i[2])
    plt.plot(notree,r2RFR)
    plt.xlabel("max depth")
    plt.ylabel("R2_Score")
plt.legend(no_tree)
```
